# 9 NumPyライブラリ

### ndarrayクラス(型)

「`NumPy`」(Numerical Python)は，多次元配列を基本的なデータ構造とした数値計算用のライブラリ(パッケージ)です。

Pythonの変数は「動的型付け」ですが，動的型付け変数では、データの取り扱いが簡単である一方で，他の言語に比べて処理速度が遅いという欠点があります。

CやFortranで書かれた型を固定して演算することができるコードをPythonから呼び出せるようにしたライブラリーが「`NumPy`」です。

以下で，100万までの整数を格納したNumPy配列とPythonのリストを用意し，それぞれに2をかけるための所要時間を確認します。





`````python
#NumPy配列での所要時間
import numpy as np
my_arr = np.arange(1000000)
%time for _ in range(10): my_arr2 = my_arr * 2
`````

`````python
#Pythonのリストでの所要時間
my_list = list(range(1000000))
%time for _ in range(10): my_list2 = my_list * 2
`````

以下では`NumPy`の中で`np.ndarray`という多次元配列を扱う「クラス」の基本を説明していきます。

`ndarray`(配列)は，**同じ型と大きさの要素からなる多次元の入れ物**です。
リストは異なる型の要素をもつことができましたが，`ndarray`(配列)では同じ型である必要があります。
またリストではリストの要素としてリストを持つことができ，そのリストの要素は異なっていても構いませんでした。
しかし`ndarray`(配列)では要素の数は同じである必要があります。

`ndarray`型のデータを生成するには、`numpy.array(リスト)`*のように配列にするリストを与えます。*

`````python
import numpy as np
b = [1, 2, 3]
a = np.array(b)
print(a)
print(b)
`````

`````python
# import numpy as np

b = np.array([[1, 2], [3, 4]])
print(b)
`````

`````python
# import numpy as np

c = np.array([[1, 2, 3], [4, 5, 6]])
print(c)
`````

a, b, cは，それぞれ以下の二次元の1×3(1行3列)の行列(3次元ベクトル)、2×2の行列、2×3の行列を表しています。

$$
a = \begin{pmatrix}
      1 & 2 & 3
\end{pmatrix},
\hspace{3cm}
b = \begin{pmatrix}
      1 & 2 \\
      3 & 4  
\end{pmatrix},
\hspace{3cm}
c = \begin{pmatrix}
      1 & 2 & 3 \\
      4 & 5 & 6
\end{pmatrix}
$$

`ndarray`(配列)では、異なる型の値をもつことができませんが、以下のセルを実行してみましょう。

`````python
import numpy as np

list_a = [1, 2, 3]
list_b = [1, 1.2, 3.1415]
list_c = ['abc', 1, 1.1]


array_a = np.array(list_a)
array_b = np.array(list_b)
array_c = np.array(list_c)
print(type(array_c[2]))
print('list_a :', list_a)
print('list_b :', list_b)
print('list_c :', list_c)
print('array_a:', array_a)
print('array_b:', array_b)
print('array_c:', array_c)
`````

````Python
array_b = np.array([1, 1.2, 3.1415])
````
では、整数`1`が実数`1.`に変換されています。
また
````Python
array_c = np.array(['abc', 1, 1.1])
````
では、整数`1`と実数`1.1`が文字列に変換されています。

このように型変換可能なものは自動的に変換されて型が統一されます。
明示的に型を統一する場合には、`dtype=型`を指定します。

`````python
# import numpy as np

a = np.array([1, 2, 3], dtype=float)
b = np.array([1, 1.2, 3.1415], dtype=int)
c = np.array(['abc', 1, 1.1])
print(a)
print(b)
print(c)
`````

### 多次元配列を1次元配列へ(ravel)

`````python
a = np.array([[1, 2, 3], [4, 5, 6]])
a_ravel = a.ravel()
print(a)
print(a_ravel)
`````

### 配列の次元数(ndim)、各次元の大きさ(shape)、要素数(size)

`````python
a = np.array([1, 2])
b = np.array([[1, 2], [3, 4]])
c = np.array([[1, 2, 3], [4, 5, 6]])
d = np.array([[[1, 2], [3, 4]], [[5, 6], [7,8]]])

print(a)
print('ndim :', a.ndim)
print('shape:', a.shape)
print('size :', a.size)
print('')
print(b)
print('ndim :', b.ndim)
print('shape:', b.shape)
print('size  :', b.size)
print('')
print(c)
print('ndim :', c.ndim)
print('shape:', c.shape)
print('size :', c.size)
print('')
print(d)
print('ndim :', d.ndim)
print('shape:', d.shape)
print('size :', d.size)
`````

### 配列(行列)の転置

属性`.T`や関数`.transpose()`で、転置行列を得ることができます。

`````python
a = np.array([[1, 2, 3], [4, 5, 6]])
print(a)
print(a.transpose())
`````

### 配列のリスト化

`.tolist()`メソッドで配列をリスト化したもの得ることができます。

`````python
a = np.array([[1, 2, 3], [4, 5, 6]])
print(a)
print(a.tolist())
print(a)
`````

### 条件を満たす要素の抽出

`````python
a = np.array([1, 2, 3, 4, 5, 6])
a[a > 2]
`````

`````python
a = np.array([1, 2, 3, 4, 5, 6])
a[(a > 2) & (a < 6)]
`````

`````python
a = np.array([1, 2, 3, 4, 5, 6])
a[(a % 2 ==0 ) | (a % 3 == 0)]
`````

`````python
a = np.array([[1, 2, 3], [4, 5, 6]])
a[a > 2]
`````

リストでは同様のことはできません。

`````python
a = [[1, 2, 3], [4, 5, 6]]
a[a > 2]
`````

### リストから配列を作成

`array()`の引数にリストを与えるとともに`.reshape()`メソッドでshapeを指定します。

`````python
a = np.array([1, 2, 3, 4, 5, 6]).reshape(2, 3)
print(a)
`````

`````python
a = np.array([1, 2, 3, 4, 5, 6]).reshape(3, 2)
print(a)
`````

`````python
a = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(2, 2, 2)
print(a)
`````

`````python
a = np.array([1, 2, 3, 4, 5, 6]).reshape(2, 2)
print(a)
`````

## 演習問題9-1

二次元のndarray配列`a`の0軸のあるインデックス`row`(行列でいう「行」)は、`a[row]`で得ることができる。
1軸のあるインデックス(行列でいう「列」)を得る関数`array_column(array, col)`を作成しなさい。

すなわち
````Python
import numpy as np

a = np.array([[1, 2, 3], [4, 5, 6]])

print(array_column(a, 0))
print(array_column(a, 2))
````
の実行結果が
````Python
[1 4]
[3 6]
````
となるような関数`array_column()`を作成しなさい。
